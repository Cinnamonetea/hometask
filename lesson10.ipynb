```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('F:/Data/BostonHousing.csv')
df.info()
```

```python
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()
```

```python
correlation_matrix = df.corr()
print(correlation_matrix['medv'].sort_values(ascending=False))
```

```python
df.groupby('rad')['medv'].mean().plot(kind='bar', title='Ср. стоимость по доступности к автомагистралям', ylabel='medv')
plt.show()
```

```python
plt.figure(figsize=(5, 5))
# sns.scatterplot(x='rad', y='tax', data=df)
df.groupby('rad')['tax'].mean().plot(kind='bar', title='Зависимость налога от близости к автомагистралям')
```

```python
group = df.groupby('rad').agg({'medv': 'mean', 'tax': 'mean'}).reset_index()
group.plot(x='rad', kind='bar', title='Средние значения medv и tax по rad', ylabel='Среднее значение', figsize=(6, 6))
```
