```python
import pandas as pd
import sqlalchemy as db
from sqlalchemy import create_engine, MetaData, Table, Column, Integer, Text, ForeignKey

data = pd.read_csv('F:/Data/globalterrorismdb.csv', encoding='ISO-8859-1')
```

```python
engine = db.create_engine('sqlite:///crimedb.db')
conn = engine.connect()
metadata = db.MetaData()
```

```python
type_incident = db.Table(
    'Type_Incident', metadata,
    db.Column('id', db.Integer, primary_key = True), 
    db.Column('alternative_txt', db.Text)
)

regions = db.Table(
    'Region', metadata,
    db.Column('id', db.Integer,  primary_key=True),
    db.Column('region_txt', db.Text) 
)

countries = db.Table(
    'Country', metadata,
    db.Column('id', db.Integer,  primary_key=True),
    db.Column('country_txt', db.Text),
    db.Column('region_id', db.Integer, ForeignKey('Region.id'))
)

cities = db.Table(
    'City', metadata,
    db.Column('id', db.Integer,  primary_key=True),
    db.Column('city_txt', db.Text),
    db.Column('country_id', db.Integer, ForeignKey('Country.id'))
)
```

```python
incident = db.Table(
    'Incident', metadata,
    db.Column('id', db.Integer,  primary_key=True),
    db.Column('iday', db.Integer),
    db.Column('imonth', db.Integer),
    db.Column('iyear', db.Integer),
    db.Column('city_id', db.Integer, ForeignKey('City.id')),
    db.Column('type_incident_id', db.Integer, ForeignKey('Type_Incident.id')) 
)
```

```python
metadata.create_all(engine)
```

```python
region_val = data[['region', 'region_txt']].drop_duplicates().values.tolist()
sort_region_val = sorted(region_val)
sort_region_val
```

```python
for region in sort_region_val:
    insertion = regions.insert().values(id=region[0], region_txt=region[1])
    conn.execute(insertion)
```

```python
conn.execute(db.select(regions)).fetchall()
```

```python
country_val = data[['country', 'country_txt', 'region']].drop_duplicates().values.tolist()
sort_country_val = sorted(country_val)
sort_country_val
```

```python
for contry in sort_country_val:
    insertion = countries.insert().values(id=contry[0], country_txt=contry[1], region_id=contry[2])
    conn.execute(insertion)
```

```python
conn.execute(db.select(countries)).fetchall()
```

```python
city_val = data[['city','country']].drop_duplicates().values.tolist()
#sort_city_val (city_val)
city_val
```

```python
step = 0
for city in city_val:
    insertion = cities.insert().values(id=step, city_txt=city[0], country_id=city[1])
    conn.execute(insertion)
    step += 1
```

```python
conn.execute(db.select(cities)).fetchall()
```

```python
conn.execute(db.select(type_incident)).fetchall()
```
